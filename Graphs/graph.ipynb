```python
import matplotlib.pyplot as plt
from matplotlib import rcParams, font_manager as fm

# Register custom Poppins font
fm.fontManager.addfont('/Users/vivek/Library/Fonts/Poppins-Regular.ttf')
rcParams['font.family'] = 'Poppins'

# Data
years = [
    "2014-15", "2015-16", "2016-17", "2017-18", "2018-19", 
    "2019-20", "2020-21", "2021-22", "2022-23", "2023-24", "2024-25"
]
values = [1940, 2059, 1521, 4682, 10746, 9116, 8434, 12814, 15920, 21083, 23622]

# Plot setup
plt.figure(figsize=(10, 6))
bars = plt.bar(
    years, values,
    color="#4A90E2",
    edgecolor="#2039B4",
    alpha=0.75,
    linewidth=1.5,
    width=0.7
)

# Annotate values
for bar, value in zip(bars, values):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        value + 400,
        f"{value:,}",
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='regular',
        color="#1A1A1A"
    )

# Main title and subtitle
plt.title("Indias Defense Exports", fontsize=30, fontweight='bold', loc='left', color="#111")
plt.suptitle(
    "From 1,940 Cr to 23,622 Cr  1117% Growth Since 2014",
    fontsize=13,
    y=0.848,
    x=0.33,
    color="#333"
)

# Add source text at the bottom left of the plot
plt.text(
    -0.5, -2000,
    "Source: Ministry of Defence, PIB",
    fontsize=9,
    color="#555",
    ha='left'
)

plt.ylabel("Export Value ( crore)", fontsize=11)

# Styling
plt.xticks(rotation=0, fontsize=9)
plt.yticks(fontsize=9)
plt.tick_params(axis='both', which='major', length=0)
plt.grid(False)
plt.box(False)
plt.tight_layout()

# Save as high-res image
plt.savefig('/Users/vivek/Code/defense_exports.png', dpi=400, bbox_inches='tight')
plt.show()
```
